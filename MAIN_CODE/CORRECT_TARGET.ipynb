```python
import os
import numpy as np

base_dir = "F:/CARLA_BRAKE/SIM_DATA/BACKUP_1"

# Scorri ogni cartella nel dataset
for folder in os.listdir(base_dir):
    folder_path = os.path.join(base_dir, folder)
    if os.path.isdir(folder_path):
        # Conta le immagini
        images = [f for f in os.listdir(folder_path) if f.endswith('.png')]
        num_images = len(images)
        
        # Percorso del file target.npy
        target_file = os.path.join(folder_path, 'target.npy')
        
        if os.path.exists(target_file):
            # Carica i target
            targets = np.load(target_file)
            num_targets = len(targets)
            
            # Verifica e aggiorna se necessario
            if num_images != num_targets:
                print(f"Incongruenza in {folder}: {num_images} immagini, {num_targets} target")
                
                # Aggiungi target mancanti con un valore predefinito (es. 0)
                if num_images > num_targets:
                    diff = num_images - num_targets
                    print(f"Aggiungo {diff} target mancanti con valore di default.")
                    targets = np.concatenate([targets, [0] * diff])
                
                # Rimuovi target in eccesso
                elif num_images < num_targets:
                    diff = num_targets - num_images
                    print(f"Rimuovo {diff} target in eccesso.")
                    targets = targets[:num_images]
                
                # Salva il file aggiornato
                np.save(target_file, targets)
                print(f"File target.npy aggiornato per {folder}.")
            else:
                print(f"Nessuna incongruenza in {folder}.")
        else:
            print(f"Manca il file target.npy in {folder}.")
```
